```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('/Users/nathandhawale/Documents/CodingDojo/Coding/Data-Enrichment/Medical.csv')
df.info()
df.head()
```

Vitamin D levels are continuous and is not skewed either direction. This is mesokurtic.

```python
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature, kde = True)
ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, color = 'green', label = f'Median = {median:,.2f}')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend()
```

Doc visits are discrete and is positively skewed. This is Leptokurtic.

```python
col1 = 'Doc_visits'
feature1 = df[col1]
mean1 = feature1.mean()
median1 = feature1.median()
std1 = feature1.std()
plus_one_std1 = mean1 + std1
minus_one_std1 = mean1 - std1
fig,ax1 = plt.subplots(figsize=(10,6))
sns.histplot(feature1, kde = True)
ax1.axvline(mean1, color = 'red',label=f'Mean = {mean1:,.2f}')
ax1.axvline(median1, color = 'green', label = f'Median = {median1:,.2f}')
ax1.axvline(plus_one_std1, color = 'black',label=f'+1 std = {plus_one_std1:,.2f}')
ax1.axvline(minus_one_std1, color = 'black', label = f'-1 std = {minus_one_std1:,.2f}')
ax1.axvspan(plus_one_std1, minus_one_std1, color = 'yellow', zorder = 0)
ax1.set_title(f'{col1}')
ax1.legend()
```
